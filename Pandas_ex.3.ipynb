```python
import os

# Change directory Pandas
os.chdir('/Users/martaspizzichino/Desktop/Pandas')

# Verify that the Directory was properly changed
print(os.getcwd())  # Questo dovrebbe restituire /Users/martaspizzichino/Desktop/Pandas

# File loans.csv
import pandas as pd
df = pd.read_csv('loans.csv')

# Visualizza le prime 5 righe del DataFrame
print(df.head())
```

```python
# List of languages of interest
languages = ['English', 'Spanish']

# Filter the dataframe to include only rows where:
# - 'original_language' is in the list of 'languages'
# - 'funded_amount' is greater than 1000
df_filtered = df[(df['original_language'].isin(languages)) & (df['funded_amount'] > 1000)]

# Display the first 5 rows of the filtered dataframe
print(df_filtered.head())
```

```python
# Sort the dataframe by 'funded_amount' in descending order
# and by 'original_language' in ascending order
df_sorted = df.sort_values(by=['funded_amount', 'original_language'], ascending=[False, True])

# Display the first 5 rows of the sorted dataframe
print(df_sorted.head())
```

```python
# Group by 'sector_name' and calculate the sum of 'funded_amount' for each sector
df_grouped = df.groupby('sector_name')['funded_amount'].sum().reset_index()

# Display the grouped data
print(df_grouped)
```
